```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("C:/DataSets/wine.data")
```

```python
predictores_columnas = ['14.23', '1.71', '2.43', '15.6', '127', '2.8', '3.06', '.28', '2.29', '5.64', '1.04', '3.92', '1065']
```

```python
objetivo_columna = ['1']
```

```python
predictores = data[predictores_columnas]
objetivo = data[objetivo_columna]
```

```python
X_train, X_test, y_train, y_test = train_test_split(predictores, objetivo, test_size=0.38, random_state=5)
```

```python
arbol = DecisionTreeClassifier()
```

```python
arboldefinicion = arbol.fit(X_train, y_train)
```

```python
plot_tree(arboldefinicion)
```

```python
prediccion_vino = arboldefinicion.predict(X_test)
```

```python
pd.crosstab(np.array([y[0] for y in y_test.values.tolist()]), prediccion_vino, rownames=["Actual"], colnames=['Predicciones']) # EXAMINA LA MATRIZ DE CONFUSION
```

```python
precision_modelo = accuracy_score(y_test, prediccion_vino)
```
